```python
import os
import random
import wget
import datetime
import matplotlib.pyplot as plt
import tensorflow as tf
from os import getcwd
from shutil import copyfile
from tensorflow.keras import layers
from tensorflow.keras import Model
from os import getcwd
from tensorflow.keras.applications import EfficientNetB4
from tensorflow.keras.optimizers import Adam
from tensorflow.keras.preprocessing.image import ImageDataGenerator
from tensorflow.keras.callbacks import TensorBoard, LearningRateScheduler, ModelCheckpoint
from keras.models import load_model
from ipywidgets import FileUpload
from IPython.display import display
from keras.preprocessing import image
```

```python
gpu_options = tf.compat.v1.GPUOptions(allow_growth=True)
session = tf.compat.v1.InteractiveSession(config=tf.compat.v1.ConfigProto(gpu_options=gpu_options))
```

```python
try:
    main_dir = f'{getcwd()}/processed_data/'
    
    train_dir = os.path.join(main_dir, "training")
    val_dir = os.path.join(main_dir, "val")
    test_dir = os.path.join(main_dir, "testing")
    
    cardboard_train = os.path.join(train_dir, "cardboard")
    glass_train = os.path.join(train_dir, "glass")
    metal_train = os.path.join(train_dir, "metal")
    paper_train = os.path.join(train_dir, "paper")
    plastic_train = os.path.join(train_dir, "plastic")
    trash_train = os.path.join(train_dir, "trash")
    
    cardboard_val = os.path.join(val_dir, "cardboard")
    glass_val = os.path.join(val_dir, "glass")
    metal_val = os.path.join(val_dir, "metal")
    paper_val = os.path.join(val_dir, "paper")
    plastic_val = os.path.join(val_dir, "plastic")
    trash_val = os.path.join(val_dir, "trash")
    
    cardboard_test = os.path.join(test_dir, "cardboard")
    glass_test = os.path.join(test_dir, "glass")
    metal_test = os.path.join(test_dir, "metal")
    paper_test = os.path.join(test_dir, "paper")
    plastic_test = os.path.join(test_dir, "plastic")
    trash_test = os.path.join(test_dir, "trash")
    
    os.mkdir(main_dir)
    
    os.mkdir(train_dir)
    os.mkdir(val_dir)
    os.mkdir(test_dir)
    
    os.mkdir(cardboard_train)
    os.mkdir(glass_train)
    os.mkdir(metal_train)
    os.mkdir(paper_train)
    os.mkdir(plastic_train)
    os.mkdir(trash_train)
    
    os.mkdir(cardboard_val)
    os.mkdir(glass_val)
    os.mkdir(metal_val)
    os.mkdir(paper_val)
    os.mkdir(plastic_val)
    os.mkdir(trash_val)
    
    os.mkdir(cardboard_test)
    os.mkdir(glass_test)
    os.mkdir(metal_test)
    os.mkdir(paper_test)
    os.mkdir(plastic_test)
    os.mkdir(trash_test)
except OSError:
    pass
```

```python
def split_data(SOURCE, TRAINING, VALIDATION, TESTING, tr_size, val_size, tst_size):
    
    garbage = os.listdir(SOURCE)
    
    list = []
    for file in garbage:
        PATH = os.path.join(SOURCE, file)
        if os.path.getsize(PATH) > 0.0:
            list.append(file) 
    
    tr_count = int(len(list)*tr_size)
    train_garbage = random.sample(list, tr_count)
    for file in train_garbage:
        copyfile(f"{SOURCE}/{file}", f"{TRAINING}/{file}") 
        
    val_count = int(len(list)*val_size)    
    val_garbage = random.sample(list, val_count)
    for file in val_garbage:
        copyfile(f"{SOURCE}/{file}", f"{VALIDATION}/{file}") 
    
    tst_count = (len(list) - (tr_count+val_count))    
    test_garbage = random.sample(list, tst_count)
    for file in test_garbage:
        copyfile(f"{SOURCE}/{file}", f"{TESTING}/{file}")
```

```python
CARDBOARD_SOURCE_DIR = f'{getcwd()}/data/cardboard'
TRAINING_CARDBOARD_DIR = cardboard_train
VAL_CARDBOARD_DIR = cardboard_val
TESTING_CARDBOARD_DIR = cardboard_test

GLASS_SOURCE_DIR = f'{getcwd()}/data/glass'
TRAINING_GLASS_DIR = glass_train
VAL_GLASS_DIR = glass_val
TESTING_GLASS_DIR = glass_test

METAL_SOURCE_DIR = f'{getcwd()}/data/metal'
TRAINING_METAL_DIR = metal_train
VAL_METAL_DIR = metal_val
TESTING_METAL_DIR = metal_test

PAPER_SOURCE_DIR = f'{getcwd()}/data/paper'
TRAINING_PAPER_DIR = paper_train
VAL_PAPER_DIR = paper_val
TESTING_PAPER_DIR = paper_test

PLASTIC_SOURCE_DIR = f'{getcwd()}/data/plastic'
TRAINING_PLASTIC_DIR = plastic_train
VAL_PLASTIC_DIR = plastic_val
TESTING_PLASTIC_DIR = plastic_test

TRASH_SOURCE_DIR = f'{getcwd()}/data/trash'
TRAINING_TRASH_DIR = trash_train
VAL_TRASH_DIR = trash_val
TESTING_TRASH_DIR = trash_test


tr_size = .7
val_size = .2
tst_size = .1

split_data(CARDBOARD_SOURCE_DIR, TRAINING_CARDBOARD_DIR, VAL_CARDBOARD_DIR, TESTING_CARDBOARD_DIR, tr_size , val_size, tst_size)
split_data(GLASS_SOURCE_DIR, TRAINING_GLASS_DIR, VAL_GLASS_DIR, TESTING_GLASS_DIR, tr_size , val_size, tst_size)
split_data(METAL_SOURCE_DIR, TRAINING_METAL_DIR, VAL_METAL_DIR, TESTING_METAL_DIR, tr_size , val_size, tst_size)
split_data(PAPER_SOURCE_DIR, TRAINING_PAPER_DIR, VAL_PAPER_DIR, TESTING_PAPER_DIR, tr_size , val_size, tst_size)
split_data(PLASTIC_SOURCE_DIR, TRAINING_PLASTIC_DIR, VAL_PLASTIC_DIR, TESTING_PLASTIC_DIR, tr_size , val_size, tst_size)
split_data(TRASH_SOURCE_DIR, TRAINING_TRASH_DIR, VAL_TRASH_DIR, TESTING_TRASH_DIR, tr_size , val_size, tst_size)
```

```python
TRAINING_DIR = train_dir                    
train_datagen = ImageDataGenerator( rescale = 1.0/255.,
                                   rotation_range=40,
                                   width_shift_range=0.2,
                                   height_shift_range=0.2,
                                   shear_range=0.2,
                                   zoom_range=0.2,
                                   horizontal_flip=True,
                                   fill_mode='nearest') 


# TRAIN GENERATOR.
train_generator = train_datagen.flow_from_directory(
    TRAINING_DIR,  
    target_size=(150, 150),  
    batch_size=10,
    class_mode='categorical')                   

VALIDATION_DIR = val_dir                            
validation_datagen = ImageDataGenerator(rescale = 1.0/255.)         

 
# VALIDATION GENERATOR.
validation_generator = validation_datagen.flow_from_directory(VALIDATION_DIR,
                                                               batch_size = 10,
                                                               class_mode = 'categorical',
                                                               target_size = (150, 150)) 

TEST_DIR = test_dir                            
test_datagen = ImageDataGenerator(rescale = 1.0/255.)         

 
# TEST GENERATOR.
test_generator = test_datagen.flow_from_directory(TEST_DIR,
                                                  batch_size = 10,
                                                  class_mode = 'categorical',
                                                  target_size = (150, 150))
```

```python
path_inception = f"{getcwd()}/inception_v3/inception_v3_weights_tf_dim_ordering_tf_kernels_notop.h5"

# Import the inception model  
from tensorflow.keras.applications.inception_v3 import InceptionV3

local_weights_file = path_inception

pre_trained_model = InceptionV3(input_shape = (150, 150, 3), 
                                include_top = False, 
                                weights = None)

pre_trained_model.load_weights(local_weights_file)

for layer in pre_trained_model.layers:
    layer.trainable = False                  

pre_trained_model.summary()
```

```python
last_layer = pre_trained_model.get_layer('mixed7')
print('last layer output shape: ', last_layer.output_shape)
last_output = last_layer.output
```

```python
from tensorflow.keras.optimizers import RMSprop

x = layers.Flatten()(last_output)
x = layers.Dense(1024, activation='relu')(x)
x = layers.Dropout(0.2)(x)                  
x = layers.Dense(6, activation='softmax')(x)           

model = Model(pre_trained_model.input, x) 

model.compile(optimizer = RMSprop(learning_rate=1e-4),
              loss = 'categorical_crossentropy',
              metrics = ['acc'] )

model.summary()
```

```python
# Train the Model
history = model.fit_generator(train_generator,
                              validation_data=validation_generator,
                              epochs=50,
                              verbose = 1)
```

```python
%matplotlib inline
import matplotlib.pyplot as plt
acc = history.history['acc']
val_acc = history.history['val_acc']
loss = history.history['loss']
val_loss = history.history['val_loss']

epochs = range(len(acc))

plt.plot(epochs, acc, 'bo', label='Training accuracy')
plt.plot(epochs, val_acc, 'r', label='Validation accuracy')
plt.title('Training and validation accuracy')
plt.legend()
plt.figure()

plt.plot(epochs, loss, 'bo', label='Training Loss')
plt.plot(epochs, val_loss, 'r', label='Validation Loss')
plt.title('Training and validation loss')
plt.legend()

plt.show()
```
